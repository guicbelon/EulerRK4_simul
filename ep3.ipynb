```python
import numpy as np

def conjugate_gradient(A, b, omega=1.85, tolerance=0.01, max_iterations=1000):
    n = len(b)
    x = np.zeros(n)  # Comeando com uma soluo inicial de zero
    r = b - A.dot(x)  # Resduo inicial
    p = r.copy()      # Vetor de direo inicial

    for k in range(max_iterations):
        alpha = np.inner(r, r) / np.inner(p, A.dot(p))
        x = x + omega * alpha * p
        r_new = r - alpha * A.dot(p)

        if np.linalg.norm(r_new) < tolerance:
            break
        
        beta = np.inner(r_new, r_new) / np.inner(r, r)
        p = r_new + beta * p
        r = r_new

    return x

# Exemplo de uso
A = np.array([[4, -1, 0],
              [-1, 4, -1],
              [0, -1, 3]])

b = np.array([2, -3, 4])

x = conjugate_gradient(A, b, omega=1.85, tolerance=0.01)
print("Soluo aproximada:", x)
```
