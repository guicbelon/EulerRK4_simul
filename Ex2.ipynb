### Exercício Computacional - PMR3401
#### Parte 2 - Método de Diferenças Finitas (MDF)

```python
# Importando o módulo da biblioteca
from meccomp_lib.ex_2 import *

# Definindo alguns parâmetros globais
DELTA = 0.02
ALLOW_PRINT = True
```

#### Análise para 15 graus

```python
# Definindo a malha para 15 graus
mesh_15 = Mesh(delta=DELTA, 
               rotation_angle_in_degrees=15,
               allow_print=ALLOW_PRINT)

# Exibindo o formato da malha
mesh_15.plot_mesh()

# Simulando a malha:
mesh_15.simul()
```

##### Apresentando as imagens geradas para a malha rotacionada 15 graus

```python
# Linha de corrente
mesh_15.plot_streamline()

# Vetor de velocidade
mesh_15.plot_velocities_vectors()

# Velocidades absolutas
mesh_15.plot_velocities()

# Pressão
mesh_15.plot_pressure(plot_contour=True)
mesh_15.plot_pressure(plot_contour=False)
```

```python
# Cálculo das forças de arrasto e sustentação
mesh_15.calculate_forces()

# Força de arrasto
print(f"Força de arrasto: {mesh_15.drag_forces}")

# Força de sustentação
print(f"Força de sustentação: {mesh_15.lifting_forces}")
```

```python
# Cálculo das forças para velocidade de 10m/s
mesh_15_10 = Mesh(delta=0.05, 
                  rotation_angle_in_degrees=15,
                  U_infinity=10,
                  allow_print=ALLOW_PRINT)

# Novas simulações
mesh_15_10.simul()
mesh_15_10.calculate_forces() 
```

```python
# Força de arrasto
print(f"Nova força de arrasto para velocidade de 10m/s: {mesh_15_10.drag_forces}")

# Força de sustentação
print(f"Nova força de sustentação para velocidade de 10m/s: {mesh_15_10.lifting_forces}")
```
